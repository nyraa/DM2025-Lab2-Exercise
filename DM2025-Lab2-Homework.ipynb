## 1. Preprocessing Steps

````python
### Add the code related to the preprocessing steps in cells inside this section
import pandas as pd
import json

with open('data/final_posts.json', 'r') as f:
    posts_data = json.load(f)

post_df = pd.DataFrame([post['root']['_source']['post'] for post in posts_data]).set_index('post_id')
data_ident_df = pd.read_csv('data/data_identification.csv')
emotion_df = pd.read_csv('data/emotion.csv')
````

````python
df = post_df.join(
    [
        data_ident_df.set_index('id')['split'],
        emotion_df.set_index('id')['emotion']
    ],
    how='left'
)
train_df = df[df['split'] == 'train'].reset_index().drop(columns=['split'])
test_df = df[df['split'] == 'test'].reset_index().drop(columns=['split', 'emotion'])
````

````python
# train_df = train_df.sample(frac=0.2, random_state=42).reset_index(drop=True)
````

````python
train_df.head()
````

````python
# remove mentions, hashtags
import re
def clean_text(text):
    text = text.strip()
    text = re.sub(r'##+', '#', text)  # replace multiple # with single #
    # text = re.sub(r'(#\w*\s?)+$', '', text)  # remove trailling hashtags
    text = re.sub(r'#(\w*)', r'\1', text)  # remove # but keep the word
    text = re.sub(r'@\w+', '', text)  # remove mentions
    text = re.sub(r'http\S+', '', text)  # remove URLs
    text = re.sub(r'\s+', ' ', text).strip()  # remove extra spaces
    return text

train_df['cleaned_text'] = train_df['text'].apply(clean_text)
test_df['cleaned_text'] = test_df['text'].apply(clean_text)
````

## 3. Model Implementation Steps

````python
### Add the code related to the model implementation steps in cells inside this section
from datasets import Dataset, DatasetDict
from sklearn.metrics import accuracy_score, f1_score
from sklearn.preprocessing import LabelEncoder
from transformers import (
    BertTokenizerFast,
    BertForSequenceClassification,
    TrainingArguments,
    Trainer,
)
````

````python
le = LabelEncoder()
le.fit(emotion_df['emotion'])

train_df['labels'] = le.transform(train_df['emotion'])
dataset = Dataset.from_pandas(train_df[['cleaned_text', 'labels']])
dataset = dataset.train_test_split(test_size=0.2)
````

````python
model_name = 'bert-base-uncased'
tokenizer = BertTokenizerFast.from_pretrained(model_name)
def tokenize_function(examples):
    return tokenizer(examples['cleaned_text'], padding='max_length', truncation=True)

tokenized_ds = dataset.map(tokenize_function, batched=True)
ds_tokenized = tokenized_ds.remove_columns(
    [c for c in tokenized_ds["train"].column_names if c not in ["input_ids", "attention_mask", "labels"]]
)
ds_tokenized.set_format(
    type='torch',
    columns=['input_ids', 'attention_mask', 'labels']
)
````

````python
model = BertForSequenceClassification.from_pretrained(
    model_name,
    num_labels=len(le.classes_),
    id2label={i: label for i, label in enumerate(le.classes_)},
    label2id={label: i for i, label in enumerate(le.classes_)},
)
training_args = TrainingArguments(
    output_dir="./bert_results",
    learning_rate=2e-5,
    per_device_train_batch_size=64,
    per_device_eval_batch_size=64,
    num_train_epochs=3,
    weight_decay=0.01,
    eval_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    metric_for_best_model="f1_weighted",
    logging_steps=50,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_ds["train"],
    eval_dataset=tokenized_ds["test"],
    compute_metrics=lambda p: {
        "accuracy": accuracy_score(p.label_ids, p.predictions.argmax(-1)),
        "f1_weighted": f1_score(p.label_ids, p.predictions.argmax(-1), average="weighted"),
    },
)
trainer.train()
````

````python
import torch

id2label = model.config.id2label
def predict_emotion(text: str) -> str:
    enc = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        padding=True,
        max_length=128
    ).to(trainer.args.device)
    model.eval()
    with torch.no_grad():
        logits = model(**enc).logits
    pred = logits.argmax(-1).item()
    return id2label[pred]

test_df['emotion'] = test_df['cleaned_text'].apply(predict_emotion)
result_df = test_df[['post_id', 'emotion']]
# rename post_id to id
result_df = result_df.rename(columns={'post_id': 'id'})
result_df.to_csv('submission/emotion_predictions.csv', index=False)
````
